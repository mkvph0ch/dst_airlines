```python
import requests
import pandas as pd
import globals
import datetime
```

```python
globals.initialize()

my_params = {'api_key': globals.airlabs_token}

method = 'flights'
api_base = 'http://airlabs.co/api/v9/'
```

```python
r = requests.get(api_base + method, params = my_params).json()
resp_list = r.get('response')
```

```python
df = pd.json_normalize(resp_list)
df.head()
```
